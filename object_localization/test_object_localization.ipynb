```python
from object_localization_test import localize_objects,get_images
imgfile_L = get_images("./layyard/*")


import matplotlib.pyplot as plt
from PIL import Image



def show_images(img,title):
        plt.subplots(figsize=(10,6))
        plt.title(title)
        #plt.text(0,10,"text")
        t = Image.open(img)
        plt.imshow(t)
```

```python
kinds = []

for val in imgfile_L:
    title = localize_objects(val)
    kinds.extend(title)
    if title ==[]:
        title = ["can't detect"]
    title = ",".join(title)
    title = "gcp result is [" + title + "]"
    # print(title)
    show_images(val,title)
```
